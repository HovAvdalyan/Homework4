```python
import urllib2
from lxml import html
from lxml.html import document_fromstring
import json

url ='https://www.youtube.com/feed/trending'
response = urllib2.urlopen(url)
page = response.read()
tree = document_fromstring(page)
```

```python
from pprint import pprint
my_list=tree.xpath('//ul[@class="expanded-shelf-content-list has-multiple-items"]')
trending=[]
for i in range(len(my_list)):
    x=my_list[i].cssselect('h3')
    y=my_list[i].cssselect('[class="yt-lockup-byline "]')
    z=my_list[i].cssselect('[class="yt-lockup-meta-info"]')
    for j in range(len(x)):
        data={}
        link=x[j].cssselect('a')[0].get('href')
        title=x[j].cssselect('a')[0].text_content()
        dur=str(x[j].cssselect('span')[0].text_content())
        user=y[j].cssselect('a')[0].text_content()
        views=z[j].cssselect('li')[1].text_content()
        data['link'] = "https://www.youtube.com"+str(link)
        data['video'] = title
        data['duration'] = dur[12:-1]
        data['views']  = views
        trending.append(data)
pprint(trending)
with open('trending_videos_xml.json', 'w') as output:
    json.dump(trending, output, sort_keys = True, indent = 4)
```

```python
import unicodecsv as csv

with open('trending_videos_xml.csv', 'w') as csv_file:
    for i in trending:
            writer = csv.writer(csv_file)
            for key, value in i.items():
                writer.writerow([key, value])
            writer.writerow([])
```
