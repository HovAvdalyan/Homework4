```python
import requests
from BeautifulSoup import *
import json
url = 'https://www.youtube.com/feed/trending'
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page)
```

```python
from pprint import pprint
my_list=soup.findAll('ul', attrs={'class':"expanded-shelf-content-list has-multiple-items"})
trending=[]
for i in range(len(my_list)):
    x=my_list[i].findAll('h3')
    y=my_list[i].findAll('div', attrs={'class':"yt-lockup-byline "})
    z=my_list[i].findAll('ul',attrs={'class':"yt-lockup-meta-info"})
    for j in range(len(x)):
        data={}
        link=x[j].find('a').get('href')
        title=x[j].find('a').get('title')
        dur=str(x[j].find('span').text)
        user=y[j].find('a').text
        views=z[j].findAll('li')[1].text
        data['link'] = "https://www.youtube.com"+str(link)
        data['video'] = title
        data['duration'] = dur[11:-1]
        data['user_name'] = user
        data['views']  = views
        trending.append(data)
pprint(trending)
with open('trending_videos.json', 'w') as output:
    json.dump(trending, output, sort_keys = True, indent = 4)
```

```python
import unicodecsv as csv

with open('trending_videos.csv', 'w') as csv_file:
    for i in trending:
            writer = csv.writer(csv_file)
            for key, value in i.items():
                writer.writerow([key, value])
            writer.writerow([])
```
